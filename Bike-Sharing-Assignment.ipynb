```python
y_train = bike_train.pop('cnt')
X_train = bike_train
```

```python
# Importing Linear Regression and RFE

from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE
```

```python
# Applying RFE

lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm, 15)
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
# Features supported by RFE

X_train.columns[rfe.support_]
```

```python
# Creating a new dataframe with RFE Supported features

col = X_train.columns[rfe.support_]

X_train_rfe = X_train[col]

X_train_rfe.head()
```

```python
#Building 1st linear model

import statsmodels.api as sm
X_train_lm1 = sm.add_constant(X_train_rfe)

lr_1 = sm.OLS(y_train, X_train_lm1).fit()

print(lr_1.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs

vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Model Evaluation

#### Residual Analysis of Training Data

```python
y_train_pred = lr_7.predict(X_train_lm7)
```

```python
res = y_train - y_train_pred
sns.distplot(res,bins=20)
plt.title('Error Terms', fontsize = 14) 
plt.xlabel('Errors', fontsize = 12)
plt.show()
```

##### Error terms are normally distribured here with mean 0.

```python
res = y_train - y_train_pred
plt.scatter(x = y_train,y = res)
plt.show()
```

##### No visible pattern in error terms

### Homoscedasticity 

```python
c = [i for i in range(0,len(X_train),1)]
plt.scatter(c,y_train-y_train_pred)
plt.title('Error Terms', fontsize = 15)
plt.xlabel('Index')
plt.ylabel('y_train-y_pred')
plt.show()
```

##### Error terms are independent of each other

```python
c = [i for i in range(0,len(X_train),1)]
plt.figure(figsize = (8,5))
plt.plot(c,y_train,color = 'r')
plt.plot(c,y_train_pred,color = 'b')
plt.show()
```

##### Actual and Predicted values follows the same pattern
